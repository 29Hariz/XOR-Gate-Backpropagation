```python
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
```

```python
inputs = np.array([[0,0],
                   [0,1],
                   [1,0],
                   [1,1]])

targets = np.array([0,1,1,0])
```

```python
model = keras.models.Sequential([
    keras.layers.Dense(units=2,activation='linear',input_shape=(2,)),
    keras.layers.Dense(units=2,activation='sigmoid'),
    keras.layers.Dense(units=1,activation='linear'),
    ]
)

model.summary()
```

```python
sgd = keras.optimizers.SGD(learning_rate=0.5)
model.compile(optimizer=sgd, loss='mse')
```

```python
outputs = model.fit(inputs,
                    targets,
                    epochs=500)

loss_values = outputs.history['loss']

plt.plot(loss_values)
plt.title('Training Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.show()
```

```python
predict = model.predict(inputs)
print(["%.3f" % x for x in np.nditer(predict)])
```

```python
weights = model.weights
for weight in weights:
    weight_values = weight.numpy()
    print(weight.name, weight_values, sep="\n", end="\n\n")
```
